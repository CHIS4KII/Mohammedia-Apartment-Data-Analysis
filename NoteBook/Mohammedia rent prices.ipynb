```python
df = pd.read_csv('rent_price_Mohammedia.csv')
df
```

```python
df.isnull().sum
```

```python
df = df.dropna().reset_index(drop=True)
df
```

```python
df['cost'] = df['cost'].replace('Prix à consulter' , None)
df['cost'] = df['cost'].str.replace('\xa0', '').str.replace(' DH', '').str.replace(' ', '').astype(float)
```

```python
df
```

```python
new_df = df.groupby('location')['cost'].agg(['sum' , 'mean' , 'min' , 'max']).reset_index()
new_df
```

```python
plt.figure(figsize=(10, 8))
colors = ['lightcoral' if mean > 5000 else 'skyblue' for mean in new_df['mean']]
plt.barh(new_df['location'], new_df['mean'], xerr=[new_df['mean'] - new_df['min'], new_df['max'] - new_df['mean']], capsize=5, color=colors)
plt.xlabel('cost')
plt.ylabel('location')
plt.title('Mean Cost with Min-Max Range per City')
plt.show()
```
